```python
import pandas as pd

# Assuming data is your DataFrame

# 1. Filter out rows where transaction_type is ONLINE
data_filtered = data[data['transaction_type'] != 'ONLINE']

# 2. Group by 'ID', 'merch_category', 'merch_postal_code'
grouped_columns = ['ID', 'merch_category', 'merch_postal_code']
grouped_data = data_filtered.groupby(grouped_columns).size().reset_index()

# 3. Group by 'merch_category', 'merch_postal_code' and create a new column 'total_merchants' which is the sum of merchants for each group
grouped_data = grouped_data.groupby(['merch_category', 'merch_postal_code']).size().reset_index(name='total_merchants')

print(grouped_data)


```

```python
# Filter rows where 'merch_postal_code' starts with '70'
brasil_data = grouped_data[grouped_data['merch_postal_code'].astype(str).str.startswith('70')]
brasil_data
```

```python
# Create the mapping dictionary
category_to_age_group = {
    'Airlines': '30-40',
    'Bars/Discotheques': '20-30',
    'Computer Network/Information Services': '30-40',
    'General Retail Stores': '30-40',
    'Grocery Stores/Supermarkets': '30-40',
    'Utilities: Electric, Gas, Water': '30-40',
    'Drug Stores/Pharmacies': '40-50',
    'Hospitals': '40-50',
    'Hotels/Motels': '20-30',
    'Restaurants': '20-30'
}

# Add a new column 'age_group' based on the mapping
brasil_data['age_group'] = brasil_data['merch_category'].map(category_to_age_group)
```

```python
brasil_data.groupby(['age_group', 'merch_postal_code'])['total_merchants'].sum().reset_index().sort_values(by='merch_postal_code', ascending=False)
```
